```python
# Import required libraries

import torch
import torchaudio

# For manipulating directory paths
from os import path, walk

# For working with datasets
# import pandas as pd

# Import soundfile for opening audio:
import soundfile as sf

# Plotting library
import matplotlib.pyplot as plt
# To embed plots within the notebook
%matplotlib inline

# To use regex for file names
import re

# Scientific and vector computation for Python
import numpy as np

# API for display tools in Python
from IPython.display import Audio, display # display for playing audio

fs = 16000
```

```python
snd_path = 'C:/Users/Gebruiker/Documents/Artificial Intelligence/B3/S3-4 Bachelor Thesis/samples_nosil'
snd_name = 'speaker_14.flac'
sample_length_secs = 0.5

for i in range(40):
    snd_name = 'speaker_'+str(i+1).zfill(2)+'.flac'
    waveform, fs = sf.read(path.join(snd_path,snd_name))
    if i == 13:
        left = waveform[:1500000]
        right = waveform[2600000:]
        waveform = np.concatenate((left,right))
    j = 0
    while int(j*fs*sample_length_secs) + int(fs*sample_length_secs) < len(waveform):
        sample_start = int(j*fs*sample_length_secs)
        sample_end = sample_start + int(fs*sample_length_secs)
        sf.write('speaker_'+str(i+1).zfill(2)+'_sample_'+str(j+1).zfill(3)+'.flac', waveform[sample_start:sample_end], fs, format='FLAC')
        j += 1
```
