```python
# Import required libraries

import torch
import torchaudio

# For manipulating directory paths
from os import path, walk

# For working with datasets
# import pandas as pd

# Import soundfile for opening audio:
import soundfile as sf

# Plotting library
import matplotlib.pyplot as plt
# To embed plots within the notebook
%matplotlib inline

# To use regex for file names
import re

# Scientific and vector computation for Python
import numpy as np

# API for display tools in Python
from IPython.display import Audio, display # display for playing audio

sample_rate = 16000
```

```python
snd_path = '/home/wessel/Documents/thesis/samples_nosil'
snd_name = 'speaker_14.flac'

for i in range(40):
    snd_name = 'speaker_'+str(i+1).zfill(2)+'.flac'
    waveform, sample_rate = sf.read(path.join(snd_path,snd_name))
    if i == 13:
        left = waveform[:1500000]
        right = waveform[2600000:]
        waveform = np.concatenate((left,right))
    for j in range(40):
        sample_start = (len(waveform)//40)*j
        sf.write('speaker_'+str(i+1).zfill(2)+'_sample_'+str(j+1).zfill(2)+'.flac', waveform[sample_start:sample_start+2*sample_rate], sample_rate, format='FLAC')
```
